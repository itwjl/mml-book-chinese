### 数学符号表

标量是小写的$$ a, b, c, \alpha, \beta, \gamma $$

向量是粗体小写$$ {\boldsymbol x, \boldsymbol y, \boldsymbol z} $$

矩阵是粗体大写$$ {\boldsymbol A, \boldsymbol B,\boldsymbol C} $$

向量或矩阵的转置$$ \boldsymbol x^\top, \boldsymbol A^\top $$

矩阵的逆$$ \boldsymbol A^{-1} $$

$ \boldsymbol x $ 和 $ \boldsymbol y $ 的内积$$ \langle \boldsymbol x, \boldsymbol y \rangle $$

$ \boldsymbol x $ 和 $ \boldsymbol y $ 的点积$$ \boldsymbol x^\top \boldsymbol y $$

（有序）元组$$ B = (\boldsymbol b_1, \boldsymbol b_2 ,\boldsymbol b_3 ) $$

水平堆叠的列向量矩阵$$ \boldsymbol B = [\boldsymbol b_1, \boldsymbol b_2 ,\boldsymbol b_3] $$

一组向量（无序）$$ \mathcal B = {\boldsymbol b_1, \boldsymbol b_2 ,\boldsymbol b_3 } $$

分别是整数和自然数$$ \mathbb{Z}, \mathbb{N} $$

分别为实数和复数$$ \mathbb{R}, \mathbb{C} $$

实数的 n 维向量空间$$ \mathbb{R}^n $$

全部量词： 对于所有 x$$ \forall x $$

存在量词： 存在 x$$ \exists x $$

a 定义为 b$$ a := b $$

b 定义为 a$$ a =: b $$

a 与 b 成正比，即 a = 常数 · b$$ a \propto b $$

复合函数: "g after f"$$ g \circ f $$

当且仅当$$ \Longleftrightarrow $$

隐含$$ \Longrightarrow $$

集合$$ \mathcal A, \mathcal C $$

a 是集合 $ \mathcal A $ 的一个元素$$ a \in \mathcal A $$

空集$$ \emptyset $$

$ \mathcal A $ 和 $ \mathcal B $ 的差集： 属于$ \mathcal A $ 但不属于$ \mathcal B $的元素集合$$ \mathcal A \setminus \mathcal B $$

维数； 由 d = 1,...,D 索引$$ D $$

数据点的数量； 由 n = 1,...,N 索引$$ N $$

大小为 m × m 的单位矩阵$$ \boldsymbol I_m $$

大小为 m × n 的零矩阵$$ \boldsymbol 0_{m,n} $$

大小为 m × n 元素都为1的矩阵$$ \boldsymbol 1_{m,n} $$

标准/规范向量（其中 i 是 1 的分量）$$ \boldsymbol e^i $$

向量空间的维数$$ \rm dim $$

矩阵 $ \boldsymbol A $ 的秩$$ \rm rk(\boldsymbol A) $$

线性映射 $ \Phi $ 的像$$ \rm Im(\Phi) $$

线性映射 $ \Phi $ 的核（零空间）$$ \rm ker(\Phi) $$

$ \boldsymbol b_1 $ 的生成子空间 (生成集合)$$ \rm span[\boldsymbol b_1] $$

矩阵 $ \boldsymbol A $ 的迹$$ \rm tr(\boldsymbol A) $$

矩阵 $ \boldsymbol A $ 的行列式$$ \rm det(\boldsymbol A) $$

绝对值或者行列式（取决于上下文）$$ \lvert {·} \rvert $$

范数; 欧几里得，除非另有说明$$ \lVert {·} \rVert $$

特征值或拉格朗日乘数$$ \lambda $$

对应于特征值 $ \lambda $ 的特征空间$$ E_\lambda $$

向量 $ \boldsymbol x $ 和 $ \boldsymbol y $ 是正交（垂直）的$$ \boldsymbol x \bot \boldsymbol y $$

向量空间$$ V $$

向量空间 V 的正交补$$ V^\bot $$

$ x_n$ 累加：$ x_1 + \ldots + x_N $$$ \sum^{N}_{n=1} {x_n} $$

$ x_n$ 累乘：$ x_1 + \ldots + x_N $$$ \prod^{N}_{n=1} {x_n} $$

参数向量$$ \boldsymbol \theta $$

$ f $ 关于 $ x $ 的偏导数$$ \frac{\partial f}{\partial x} $$

$ f $ 关于 $ x $ 的全导数$$ \frac{ {\rm d}f}{ {\rm d}x} $$

倾斜度$$ \nabla $$

$ f $ 的最小函数值$$ f_* = \min{_x}f(x) $$

使得 $ f $ 最小化的值 $ x_∗ $（注意：arg min 返回一组值）$$ x_* \in {\rm arg} , \min{_x}f(x) $$

拉格朗日$$ \mathfrak{L} $$

负对数似然$$ \mathcal{L} $$

二项式系数，$ n $ 选 $ k $$$ \left(^n_k \right) $$

方差是 $ \boldsymbol x $ 的随机变量 $ X $$$ \mathbb{V}_X[\boldsymbol x] $$

期望是 $ \boldsymbol x $ 的随机变量 $ X $$$ \mathbb{E}_X[\boldsymbol x] $$

$ x $ 和 $ y $ 之间的协方差$$ {\rm Cov}_{X,Y}[\boldsymbol x, \boldsymbol y] $$

$ X,Y $对于给定 $ Z $ 是条件独立的$$ a \perp\kern-9mu\perp b \mid c $$

随机变量 $ X $ 根据 $ p $ 分布$$ X \sim p $$

具有均值 $ \boldsymbol \mu $ 和协方差 $ \boldsymbol \Sigma $ 的高斯分布$$ \mathcal N (\boldsymbol \mu, \boldsymbol \Sigma) $$

参数为 $ \mu $ 的伯努利分布$$ {\rm Ber}(\mu) $$

参数为 $ N, \mu $ 的二项式分布$$ {\rm Bin}(N, \mu) $$

参数为 $ \alpha, \beta $ 的Beta分布$$ {\rm Beta}(\alpha, \beta) $$

### 缩略语表

e.g.	Exempli gratia（拉丁文：例如）

GMM	高斯混合模型（Gaussian mixture model）

i.e.	Id est（拉丁语：这意味着）

i.i.d.	独立同分布（Independent, identically distributed）

MAP	最大后验（Maximum a posteriori）

MLE	最大似然估计（Maximum likelihood estimation/estimator）

ONB	正交基（Orthonormal basis）

PCA	主成分分析（Principal component analysis）

PPCA	概率主成分分析（Probabilistic principal component analysis）

REF	行阶梯形矩阵（Row-echelon form）

SPD	对称，正定（Symmetric, positive definite）

SVM	支持向量机（Support vector machine）

机器学习的核心是三个概念：数据（data）、模型（model）和学习（learning）。
